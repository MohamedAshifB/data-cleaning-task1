```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Data Analyst Intern\Task 1\KaggleV2-May-2016.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=['PatientId','AppointmentID'])
```

```python
df
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace('-','_').str.replace(' ','_')
```

```python
df.columns
```

```python
df['scheduledday'] = pd.to_datetime(df['scheduledday'])
df['appointmentday'] = pd.to_datetime(df['appointmentday'])
```

```python
df['no_show'] = df['no_show'].str.strip().str.upper().map({'NO': 0, 'YES': 1})
```

```python
df = df.drop_duplicates()
```

```python
print(df.isnull().sum())
print(df.dtypes)
```

```python
df.to_csv("cleaned_medical_appointments.csv", index=False)
print("✅ Cleaning complete. File saved as 'cleaned_medical_appointments.csv'.")
```
